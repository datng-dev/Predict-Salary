### THE DATA

** Read in the census_data.csv data with pandas**

```python
import pandas as pd
```

```python
my_data = pd.read_csv('census_data.csv')
```

```python
my_data
```

** TensorFlow won't be able to understand strings as labels, you'll need to use pandas .apply() method to apply a custom function that converts them to 0s and 1s. **

** Convert the Label column to 0s and 1s instead of strings.**

```python
my_data['income_bracket'].unique()
```

```python
def label_fix(label):
    if label == ' <=50K':
        return 0
    else:
        return 1
```

```python
my_data['income_bracket'] = my_data['income_bracket'].apply(label_fix)
```

```python
my_data.head()
```

### Perform a Train Test Split on the Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_data = my_data.drop('income_bracket', axis=1)
```

```python
y_labels = my_data['income_bracket']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_labels, test_size=0.33, random_state=101)
```

### Create the Feature Columns for tf.esitmator

** Take note of categorical vs continuous values! **

```python
my_data.columns
```

** Import Tensorflow **

```python
import tensorflow as tf
```

** Create the tf.feature_columns for the categorical values. Use vocabulary lists or just use hash buckets. Use jash buckets when u dont know how many categori in this feature**

```python
workclass = tf.feature_column.categorical_column_with_hash_bucket("workclass", hash_bucket_size=1000)

education = tf.feature_column.categorical_column_with_hash_bucket("education", hash_bucket_size=1000)

marital_status = tf.feature_column.categorical_column_with_hash_bucket("marital_status", hash_bucket_size=1000)

relationship = tf.feature_column.categorical_column_with_hash_bucket("relationship", hash_bucket_size=1000)

occupation = tf.feature_column.categorical_column_with_hash_bucket("occupation", hash_bucket_size=1000)

native_country = tf.feature_column.categorical_column_with_hash_bucket("native_country", hash_bucket_size=1000)

gender = tf.feature_column.categorical_column_with_vocabulary_list("gender", ['Female', 'Male'])

race = tf.feature_column.categorical_column_with_vocabulary_list("race", ['White', 'Black'])
```

** Create the continuous feature_columns for the continuous values using numeric_column **

```python
age = tf.feature_column.numeric_column("age")

education_num = tf.feature_column.numeric_column('education_num')

capital_gain = tf.feature_column.numeric_column('capital_gain')

capital_loss = tf.feature_column.numeric_column('capital_loss')

hours_per_week = tf.feature_column.numeric_column('hours_per_week')
```

** Put all these variables into a single list with the variable name feat_cols **

### Create Input Function

** Batch_size is up to you. But do make sure to shuffle!**

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=100, num_epochs=None, shuffle=True)
```

#### Create your model with tf.estimator

**Create a LinearClassifier.(If you want to use a DNNClassifier, keep in mind you'll need to create embedded columns out of the cateogrical feature that use strings.)**

```python
embedded_workclass = tf.feature_column.embedding_column(workclass, dimension=1000)
embedded_education = tf.feature_column.embedding_column(education, dimension=1000)
embedded_marital_status= tf.feature_column.embedding_column(marital_status, dimension=1000)
embedded_relationship = tf.feature_column.embedding_column(relationship, dimension=1000)
embedded_occupation = tf.feature_column.embedding_column(occupation, dimension=1000)
embedded_native_country = tf.feature_column.embedding_column(native_country, dimension=1000)
embedded_gender = tf.feature_column.embedding_column(gender, dimension=2)
embedded_race = tf.feature_column.embedding_column(race, dimension=2)
```

```python
feat_cols = [ embedded_workclass, embedded_education, embedded_marital_status, embedded_relationship, embedded_occupation, 
             embedded_native_country, embedded_gender, embedded_race, age, education_num, capital_gain, capital_loss, 
             hours_per_week ]
```

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10,20,20,20,10], feature_columns=feat_cols, n_classes=2)
```

** Train your model on the data. **

```python
dnn_model.train(input_fn=input_func, steps=10000)
```

### Evaluation



```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(X_test, y_test, batch_size=100, num_epochs=1, shuffle=False)
```

```python
dnn_model.evaluate(eval_input_func)
```

** Use model.predict() and pass in your input function. This will produce a generator of predictions, which you can then transform into a list, with list() **

```python
pred_func = tf.estimator.inputs.pandas_input_fn(x=X_test, batch_size=len(X_test), shuffle=False)
```

```python
pred_gen = dnn_model.predict(input_fn=pred_func)
```

```python
predictions = list(pred_gen)
```

```python
final_preds = [pred['class_ids'][0] for pred in predictions]
```
